```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the rawdata.json file
df = pd.read_json('rawdata.json')

# Extract 'e' and 't' from the 'vitals' column
df['e'] = df['vitals'].apply(lambda x: x.get('e', None))
df['t'] = df['vitals'].apply(lambda x: x.get('t', None))

# Drop rows where 'e' or 't' is missing
df.dropna(subset=['e', 't'], inplace=True)

# Convert 't' values to a relative scale by subtracting the minimum timestamp
df['t'] = df['t'] - df['t'].min()

# Plot the data
plt.plot(df['t'], df['e'], marker='o')
plt.xlabel('Time (relative)')
plt.ylabel('Heartbeat (e)')
plt.title('Heartbeat Waveform')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the rawdata.json file
df = pd.read_json('rawdata.json')

# Extract 'e' and 't' from the 'vitals' column
df['e'] = df['vitals'].apply(lambda x: x.get('e', None))
df['t'] = df['vitals'].apply(lambda x: x.get('t', None))

# Drop rows where 'e' or 't' is missing
df.dropna(subset=['e', 't'], inplace=True)

# Convert 't' values to a relative scale by subtracting the minimum timestamp
df['t'] = df['t'] - df['t'].min()

# Plot the data without gridlines or background lines
plt.plot(df['t'], df['e'], color='blue', linewidth=2)
plt.xlabel('Time (relative)')
plt.ylabel('Heartbeat (e)')
plt.title('Heartbeat Waveform')

# Remove background and axes lines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

# Remove ticks
plt.xticks([])
plt.yticks([])

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the rawdata.json file
df = pd.read_json('rawdata.json')

# Extract 'e' and 't' from the 'vitals' column
df['e'] = df['vitals'].apply(lambda x: x.get('e', None))
df['t'] = df['vitals'].apply(lambda x: x.get('t', None))

# Drop rows where 'e' or 't' is missing
df.dropna(subset=['e', 't'], inplace=True)

# Convert 't' values to a relative scale by subtracting the minimum timestamp
df['t'] = df['t'] - df['t'].min()

# Plot the data
plt.plot(df['t'], df['e'], color='blue', linewidth=2)
plt.xlabel('Time (relative)', fontsize=12)
plt.ylabel('Heartbeat (e)', fontsize=12)
plt.title('Heartbeat Waveform', fontsize=14)

# Keep x and y axes but hide the top and right borders
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Keep left and bottom borders visible (x-axis and y-axis)
plt.gca().spines['left'].set_visible(True)
plt.gca().spines['bottom'].set_visible(True)

# Show ticks on x and y axis
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Show the plot
plt.show()
```
